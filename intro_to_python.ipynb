```python
players = [{
   'name': 'A.J. Brown',
   'catches': 88,
   'targets': 145
   },
   {
   'name': 'CeeDee Lamb',
   'catches': 107,
   'targets': 156
   },
   {
   'name': 'Justin Jefferson',
   'catches': 128,
   'targets': 184
   },
]

for player in players:
  name = player['name']
  catches = player['catches']
  targets = player['targets']
  catch_rate = catches/targets
  catch_percentage = str(round(catch_rate * 100, 2))
  player['catch_percentage'] = catch_percentage
  print('In 2022 {} had a {}% catch rate'.format(name, catch_percentage))

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# The code to create the players list goes here (as shown in your question)

# Convert the players list into a pandas DataFrame
df = pd.DataFrame(players)

# Convert the catch_percentage column from string to numeric (float)
df['catch_percentage'] = df['catch_percentage'].astype(float)

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(df['name'], df['catch_percentage'])
plt.xlabel('Player Name')
plt.ylabel('Catch Rate (%)')
plt.title('Catch Rate of Players in 2022')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mock data for player rushing yards over the season
data = {
    'Player': ['Ezekiel Elliott', 'Derrick Henry', 'Christian McCaffrey', 'Saquon Barkley', 'Dalvin Cook'],
    'Week1': [89, 120, 82, 68, 94],
    'Week2': [112, 98, 105, 94, 120],
    'Week3': [72, 145, 68, 112, 90],
    'Week4': [88, 122, 101, 78, 115],
    'Week5': [98, 103, 75, 128, 101],
    # Add more weeks here
}

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Melt the DataFrame to reshape it for visualization (Week and Rushing Yards as columns)
melted_df = pd.melt(df, id_vars=['Player'], var_name='Week', value_name='Rushing Yards')

# Sort the DataFrame by Player and Week to ensure proper ordering in the plot
melted_df = melted_df.sort_values(by=['Player', 'Week'])

# Set up the plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_df, x='Week', y='Rushing Yards', hue='Player', marker='o', linewidth=2)

# Customize the plot
plt.xlabel('Week')
plt.ylabel('Rushing Yards')
plt.title('NFL Player Rushing Yards Over the Season')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
